1: Mobile Recharge Validation System


```python
def validate_recharge(amount):
  valid_plans = [199, 299, 399, 599]
  if amount < 50:
    print("Recharge Amount must be atleast 50")
    return False
  elif amount not in valid_plans:
    print("Invalid plan")
    print("Available plans are:", valid_plans)
    return False
  else:
    print("Recharge successful for Rupees", amount)
    return True
while True:
  recharge_amount = int(input("Enter recharge amount:"))
  if validate_recharge(recharge_amount):
    break
  else:
    print("Please try again.\n")
```

2: Inventory Reorder Alert System


```python
def check_stock(products):
  for product, quantity in products.items():
    if quantity < 15:
      print(product, "→ Reorder Alert(Stock:", quantity, ")")
    else:
      print(product, "→ Stock OK (Stock", quantity,")")
inventory = {
    "Rice" : 20,
    "Sugar" :10,
    "Oil" : 5,
    "soap" : 25,
    "Milk" :12
}
check_stock(inventory)
```

3: Student Result Processing System


```python
def evaluate_student(marks):
  total = 0
  for mark in marks:
    total += mark
  average = total / len(marks)
  if average >= 50:
    return average, "Pass"
  else:
    return average, "Fail"
student_marks = [60, 45, 70, 55, 40]
avg, result = evaluate_student(student_marks)
print("Average Marks:", avg)
print("Result:", result)
```

 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
  base_fare = 50
  per_km_rate = 12
  total_fare = base_fare + (distance * per_km_rate)
  if peak.lower() == "yes":
    total_fare = total_fare*1.25
  return total_fare
while True:
  try:
    distance = float(input("Enter distance in km:"))
    if distance <= 0:
      print("Distance must be gretaer than 0")
      continue
    peak_hour = input("Is it peak hour? (yes/no):")
    fare = calculate_fare(distance, peak_hour)
    print("Total cab fare:", fare)
    break
  except ValueError:
    print("Please enter a valid number")
```

 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance):
  present_count = 0
  for day in attendance:
    if day == "P":
      present_count += 1
  total_days = len(attendance)
  if total_days == 0:
    return "No attendance data available"
  percentage = (present_count / total_days) * 100
  if percentage >= 75:
    return percentage, "Eligible"
  else:
    return percentage, "Not eligible"
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_eligibility(attendance_list)
print("Attendance Percentage:", result[0])
print("Eligibility:", result[1])
```

 6: Password Strength Checker

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    if len(password) < 8:
        return " Weak Password (Minimum length is 8)"
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    if has_digit and has_special:
        return " Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"
pwd = input("Enter your password: ")
result = check_password_strength(pwd)
print(result)
```
