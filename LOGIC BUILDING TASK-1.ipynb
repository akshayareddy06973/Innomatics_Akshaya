2. Pass / Fail Analyzer


```python
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
for mark in marks:
  #A student passes if marks â‰¥ 50
  if mark >= 50:
    #Counting the total number of pass students
    pass_count += 1
  else:
    #Counting the total number of fail students
    fail_count += 1
#Print the final result clearly
print("Total Students:",len(marks))
print("Passed Students:",pass_count)
print("Failed Students:",fail_count)
```

3. Simple Data Cleaner


```python
names=["Alice","bob","CHARLIE"]
cleaned_names=[]
for name in names:
  #Removing extra spaces
  #Convert all names to lowercase
  cleaned_name=name.strip().lower()
  cleaned_names.append(cleaned_name)
print("Cleaned Names:",cleaned_names)
```

4. Message Length Analyzer


```python
messages=["Hi","Welcome to the platform","OK"]
for message in messages:
  #Print the length of each message
  length=len(message)
  print("Message:",message)
  print("Length:",length)
  #Flag messages longer than 10 characters
  if length > 10:
     print("Status:Long")
  else:
     print("Status:Normal")
  print()
```

5. Error Message Detector


```python
logs=["INFO","ERROR","WARNING","ERROR"]
error_count = 0
for log in logs:
  if log == "ERROR":
    #Print the total error count
    error_count += 1
print("Cleaned Names:",error_count)
```
