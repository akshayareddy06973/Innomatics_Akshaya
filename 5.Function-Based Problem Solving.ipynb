1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
  total_likes = 0
  for likes in likes_list:
    total_likes += likes
  print("Total likes:", total_likes)
  if total_likes >= 1000:
    print("Post is popular")
  else:
    print("Post is not popular")
likes = [200, 300, 150, 400]
analyze_engagement(likes)
```

3: Agriculture – Rainfall Adequacy Checker


```python
def analyze_rainfall(rainfall_data, required_level):
  total = 0
  for rain in rainfall_data:
    total += rain
  average = total / len(rainfall_data)
  print(f"Average rainfall: {int(average)}")
  if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
  else:
        print("Rainfall Status: Inadequate Rainfall")
rainfall_values = [60, 80, 75, 70, 85, 65]
required_rainfall = 70
analyze_rainfall(rainfall_values, required_rainfall)

```

4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_usernames(usernames):
    unique_usernames = set(usernames)
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
username_list = ["alex123", "john_doe", "alex123", "maria89"]
detect_duplicate_usernames(username_list)
```

6: Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    print(f"Premium Crops: {premium}")
crop_prices = [1500, 2500, 1800, 3200, 2000]
filter_premium_crops(crop_prices)
```

 7: System Monitoring – Application Health Checker


```python
def evaluate_system_health(errors):
    print(f"Error Count: {errors}")
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")
evaluate_system_health(7)
```

8: Banking – Daily Transaction Limit Checker

```python
def validate_transaction(amount):
    daily_limit = 50000
    print(f"Transaction Amount: {amount}")
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")
validate_transaction(60000)
```

9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance_record):
    total_days = len(attendance_record)
    present_days = 0
    for day in attendance_record:
        if day == 1:
            present_days += 1
    attendance_percentage = (present_days / total_days) * 100
    print(f"Attendance Percentage: {attendance_percentage}")
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
attendance = [1, 1, 0, 1, 1]
check_exam_eligibility(attendance)
```

10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
units_consumed = 250
bill_amount, usage_status = calculate_electricity_bill(units_consumed)
print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{bill_amount}")
print(f"Usage Level: {usage_status}")
```
