 1:Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    current_parked = 0
    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            current_parked -= 1
    print("Currently Parked Vehicles:", current_parked)

    if current_parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_system(capacity, logs)
```

2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic_level, weather):
    base_speed = 20
    base_time = (distance / base_speed) * 60
    delay = 0
    if traffic_level == "High":
        delay += 21
    elif traffic_level == "Medium":
        delay += 10
    elif traffic_level == "Low":
        delay += 5
    if weather == "Rainy":
        delay += 10
    elif weather == "Storm":
        delay += 20
    elif weather == "Clear":
        delay += 0
    final_time = base_time + delay
    print("Estimated Delivery Time:", int(final_time), "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

3:Movie Theatre Seat Occupancy Analyzer


```python
def analyze_seat_booking(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open Additional Show")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
total_seats = 200
booked_seats = [1] * 150
analyze_seat_booking(total_seats, booked_seats)
```

4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average = total / len(cpu_readings)
    print("Average CPU Load:", int(average), "%")
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

5:Smart Classroom Resource Usage Monitor

```python
def track_resource_usage(resource_usage):
    overused = []
    threshold = 8
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
track_resource_usage(usage)
```

6: Online Event Registration Capacity Controller


```python
def manage_event_registration(capacity, registrations):
    confirmed = 0
    waitlisted = 0
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
manage_event_registration(100, 105)
```
