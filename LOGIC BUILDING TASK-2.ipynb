1.Unique words in a sentence

```python
sentence = input("Enter a sentence:")
#splitting into words
words=sentence.split()
#covert list to set for removing duplicates
unique_words=set(words)
ordered_words=sorted(unique_words)
#print result
print("unique words count:",len(ordered_words))
print("unique words:",ordered_words)
```

2.Highest Salary from Employee Data


```python
employees = {
    "Ravi" : 75000,
    "Anitha" : 60000,
    "Kiran" : 72000
}
#Employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]
print("Highest Salary", highest_employee,"-",highest_salary)
```

3.Find Maximum and Minimum Values


```python
numbers=[45, 22, 89, 10, 66]
#Initialize min and max with first element
maximum = numbers[0]
minimum = numbers[0]
#loop for comparing
for num in numbers:
  if num > maximum:
    maximum = num
  if num < minimum:
    minimum = num
print("List", numbers)
print("Max", maximum)
print("min", minimum)
```

4. Count Products Above a Price Threshold


```python
prices=[450, 1200, 899, 1500, 300]
count=0
for price in prices:
  if price > 1000:
    count += 1
print("Products above 1000:",count)
```

5.Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]
#count no.of P values
present_count = 0
for day in attendance:
  if day == "P":
    present_count += 1
#calculating percentage
percentage = (present_count / len(attendance)) * 100
print("Attendance Percentage:",percentage)
```

7.Count Character Frequency


```python
text = "pythonp"
char_count = {}
for ch in text:
  if ch in char_count:
    char_count[ch] += 1
  else:
    char_count[ch] = 1
print(char_count)
```
